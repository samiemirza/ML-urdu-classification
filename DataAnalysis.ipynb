```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("full.csv")
```

```python
unique_gold_labels = df['gold_label'].nunique()
print(f"Number of unique values in 'gold_label' column: {unique_gold_labels}")
```

```python
gold_label_counts = df['gold_label'].value_counts()
print("Number of occurrences for each unique value in 'gold_label':")
print(gold_label_counts)
```

```python
gold_label_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Occurrences for Each Unique Gold Label')
plt.xlabel('Gold Label')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
missing_data = df.isnull().sum()
print("Missing data in each column:")
print(missing_data)
```

```python
df.dropna(inplace=True)
remaining_nulls = df.isnull().sum().sum()
print(df.shape)
```

```python
df_unique = df.drop_duplicates(subset='content', keep='first')

print("Original DataFrame:")
print(df.shape)

print("\nDataFrame after dropping duplicates:")
print(df_unique.shape)
```

```python
df1 = df_unique[:901]
df2 = df_unique[901:1802]
df3 = df_unique[1802:]
```

```python
print(df1.shape)
print(df2.shape)

print(df3.shape)
```

```python
df1.to_csv('df1.csv', index=False)
df2.to_csv('df2.csv', index=False)
df3.to_csv('df3.csv', index=False)

print("DataFrames saved as CSV files.")
```
